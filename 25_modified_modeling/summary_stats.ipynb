```python
deaths[(deaths["STNAME"]=="Florida") & (deaths["policy_year"]==1)][["Deaths", "deaths_per_cap"]].describe()
```

```python
# difference in means

# before policy
diff1 = (
    opi[(opi["BUYER_STATE"] == "FL") & (opi["policy_year"] == 0)]["MME per cap"].mean()
    - opi[(opi["BUYER_STATE"].isin(control_fl)) & (opi["policy_year"] == 0)][
        "MME per cap"
    ].mean()
)

# after policy
diff2 = (
    opi[(opi["BUYER_STATE"] == "FL") & (opi["policy_year"] == 1)]["MME per cap"].mean()
    - opi[(opi["BUYER_STATE"].isin(control_fl)) & (opi["policy_year"] == 1)][
        "MME per cap"
    ].mean()
)

print(diff1, diff2)
```

```python
# difference in means

# change of control before and after policy
diff1 = (
    opi[(opi["BUYER_STATE"].isin(control_fl)) & (opi["policy_year"] == 1)]["MME per cap"].mean()
    - opi[(opi["BUYER_STATE"].isin(control_fl)) & (opi["policy_year"] == 0)][
        "MME per cap"
    ].mean()
)

# change of Florida before and after policy
diff2 = (
    opi[(opi["BUYER_STATE"] == "FL") & (opi["policy_year"] == 1)]["MME per cap"].mean()
    - opi[(opi["BUYER_STATE"] == "FL") & (opi["policy_year"] == 0)]["MME per cap"].mean()
)

print(diff1, diff2)
```
