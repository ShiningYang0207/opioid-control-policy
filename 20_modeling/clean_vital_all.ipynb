```python
import pandas as pd

# load 'Cause of Death' data of each year (2003-2015) and merge them into one dataframe
vital = pd.read_csv(f'/Users/Jung/Downloads/US_VitalStatistics/CauseofDeath2003.txt', sep='\t')
for i in range(2004, 2016):
    new_vital = pd.read_csv(f'/Users/Jung/Downloads/US_VitalStatistics/CauseofDeath{i}.txt', sep='\t')
    vital = pd.concat([vital, new_vital])

vital['Year'].value_counts()
```

```python
vital['Year'].value_counts()
```

```python
# column 'Notes' has uneccessary data, so delete the coulumn and related rows
print(vital['Notes'].value_counts())
vital.drop('Notes',axis = 1, inplace = True)
vital.dropna(how = 'all', inplace = True)
vital
```

```python
# extract only drug-related data
print(vital['Drug/Alcohol Induced Cause'].value_counts())
str = ['Drug poisonings (overdose) Unintentional (X40-X44)', 'Drug poisonings (overdose) Suicide (X60-X64)', \
    'Drug poisonings (overdose) Undetermined (Y10-Y14)', 'All other drug-induced causes', 'Drug poisonings (overdose) Homicide (X85)']
vital_drug = vital.loc[vital['Drug/Alcohol Induced Cause'].isin(str)]
vital_drug
```

```python
# drop 'Year Code' and 'Drug/Alcohol Induced Cause Code' columns
str = ['Year Code', 'Drug/Alcohol Induced Cause Code']
vital_drug.drop(str, axis = 1, inplace = True)
vital_drug
```

```python
# change the types of columns which have numbers from float to integer
# check if there is null data
# it looks fine, but..
vital_drug.isnull().any()
```

```python
# check the types of columns which have numbers
vital_drug.dtypes
```

```python
# type of 'Deaths' column is an object
# so, change it to numeric
vital_drug['Deaths'] = pd.to_numeric(vital_drug['Deaths'], errors = 'coerce')

# now we can find missing values
vital_drug.isnull().any()
```

```python
# total rows are 10432, but rows of 'Deaths' column are 10407 -> there are 25 missing values
vital_drug['Deaths'].describe()
```

```python
# the number of missing values is very small compared to the total number
# so, just replace them to 0
vital_drug['Deaths'] = pd.to_numeric(vital_drug['Deaths'], errors = 'coerce').fillna(0)
vital_drug
```

```python
# change the types of columns which have numbers from float to integer
str = ['County Code', 'Year', 'Deaths']
vital_drug[str] = vital_drug[str].astype(int)
vital_drug
```

```python
# now we don't need 'Drug Induced Cause' any more
vital_drug.drop('Drug/Alcohol Induced Cause', axis = 1, inplace = True)
vital_drug
```

```python
# add the number of deaths by county and year
vital_drug_group = vital_drug.groupby([vital_drug['County'], vital_drug['County Code'], vital_drug['Year']], as_index = False).sum()
vital_drug_group
```
